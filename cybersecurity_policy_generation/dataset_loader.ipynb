```python
def get_txt_files(folder: str = "Templates", recursive: bool = False) -> None:
    base = Path(folder).resolve()
    if not base.exists():
        print(f"❌ Folder not found: {base}")
        return

    files = base.rglob("*.txt") if recursive else base.glob("*.docx")
    return files
```

```python
def sha256(s: str) -> str:
    import hashlib
    return hashlib.sha256(s.encode("utf-8")).hexdigest()

def compact(s: str) -> str:
    return re.sub(r"[ \t]+", " ", s).strip()

def make_id(policy_id: str, number: str) -> str:
    safe = number.replace(" ", "").replace("•", "-")
    return f"{policy_id}:{safe}"


# Section headings (choose what you need; first that matches wins)
SECTION_RE = re.compile(
    r"""^
    \s*                                         # optional leading spaces
    (                                           # group 1: full marker (no trailing dot)
      (?:\d+|[IVXLCDM]+|[A-Z])                  # one segment: digits OR roman OR single letter
      (?:\.(?:\d+|[IVXLCDM]+|[A-Z]))*           # .segment repeated
    )
    \.                                          # REQUIRED trailing dot
    \s+                                         # at least one space before title
    (.*\S)?                                     # group 2: title (optional, trimmed later)
    \s*$                                        # optional trailing spaces to end
    """,
    re.VERBOSE
)
SECTION_RES = [
    SECTION_RE
]
# SECTION_RES = re.compile(
#     r"^((?:\d+|[IVXLCDM]+|[A-Z])(?:\.(?:\d+|[IVXLCDM]+|[A-Z]))*)\.?\s*(.*\S)?\s*$",
#     re.IGNORECASE
# )

# Numbered clauses like "3.1", "3.1.4", "2.3.1.1", with optional trailing dot then text
CLAUSE_RE = re.compile(r"^(\d+(?:\.\d+)+)\.?\s+(.*\S.*)$")

# A “list header” is a numbered clause followed by a Title/Heading (kept) and then bullets
LIST_HEADER_RE = re.compile(r"^(\d+(?:\.\d+)+)\.?\s+([A-Z].+)$")

# Bullets: "• text", "- text", "* text", "● text"

BULLET_RE = re.compile(r"^\s*[\-\*\u2022\u25CF]\s+(.*\S.*)$")

# -------------------------
# Category mapping (optional)
# -------------------------
def infer_category(number: str, section_title: str, category_map: Optional[Dict[str,str]] = None) -> str:
    """
    Fallback rule:
    - If a prefix of the clause number is in category_map (e.g., '3.2' -> 'Internet'), use that.
    - Else use a normalized section title as category.
    """
    if category_map:
        # try longest prefix match on dotted numbers, e.g., '3.5.11.1' -> '3.5.11' -> '3.5' -> '3'
        parts = number.split('.')
        for k in range(len(parts), 0, -1):
            key = '.'.join(parts[:k])
            if key in category_map:
                return category_map[key]
    return re.sub(r"\W+", "", section_title).strip() or "General"

# -------------------------
# Core parser (framework-agnostic)
# -------------------------
def parse_policy(
    text: str,
    policy_id: str,
    title: str,
    version: str,
    *,
    owner: str = "Policy Owner",
    policy_uri: Optional[str] = None,
    category_map: Optional[Dict[str,str]] = None
) -> Dict[str, Any]:
    """
    Parse ANY numbered policy into sections and clauses.
    - Sections: matched by SECTION_RES (1., I., A., …)
    - Clauses:   matched by CLAUSE_RE (1.1, 1.2.3, …)
    - List headers: numbered clause with a TitleCase trailing text; bullets under it become individual clauses
    - Bullets: each bullet line becomes a clause "X.Y • N"

    Returns a dict with Policy + sections[*].clauses[*]
    """
    lines = [l.rstrip() for l in text.splitlines()]

    # 1) Find section start indices
    sec_starts: List[int] = []
    for i, ln in enumerate(lines):
        for rx in SECTION_RES:
            if rx.match(ln):
                sec_starts.append(i)
                break
    if not sec_starts:
        # no explicit sections detected; treat whole doc as section "1"
        lines = [f"1. {title}"] + lines
        sec_starts = [0]

    sec_starts.append(len(lines))

    sections: List[Dict[str, Any]] = []

    # 2) Walk sections
    for si in range(len(sec_starts)-1):
        s_start, s_end = sec_starts[si], sec_starts[si+1]
        head = lines[s_start]
        sec_num, sec_title = _match_section(head)
        sec_body = [l for l in lines[s_start+1:s_end]]
        sec_text = "\n".join(sec_body).strip()

        sec = {
            "id": f"{policy_id}:S{sec_num}",
            "number": sec_num,
            "title": compact(sec_title),
            "text": sec_text,
            "sha256": sha256(sec_text),
            "clauses": []
        }

        # 3) Clause extraction within this section
        current_list_number = None   # e.g., "3.7.2"
        current_list_title  = None
        bullet_index = 0

        i = 0
        while i < len(sec_body):
            raw = sec_body[i].strip()
            if not raw:
                i += 1
                continue

            # a) Numbered clause (generic)
            m = CLAUSE_RE.match(raw)
            if m:
                number, rest = m.group(1), m.group(2).strip()

                # Try to detect "list header": a numbered heading followed by bullets.
                mlh = LIST_HEADER_RE.match(raw)
                if mlh and _peek_has_bullet(sec_body, i+1):
                    current_list_number = mlh.group(1)          # e.g., "3.7.2"
                    current_list_title  = compact(mlh.group(2)) # e.g., "System and Network Activities"
                    bullet_index = 0
                    # Optionally persist the list header as a heading node
                    sec["clauses"].append({
                        "id": make_id(policy_id, number),
                        "number": number,
                        "title": current_list_title,
                        "text": "",
                        "category": infer_category(number, sec_title, category_map),
                        "section_number": sec_num,
                        "path": f"{title} > {sec_num} {sec_title} > {number} {current_list_title}",
                        "sha256": sha256(""),
                        "is_heading": True
                    })
                    i += 1
                    continue

                # Regular numbered clause
                sec["clauses"].append({
                    "id": make_id(policy_id, number),
                    "number": number,
                    "title": None,
                    "text": rest,
                    "category": infer_category(number, sec_title, category_map),
                    "section_number": sec_num,
                    "path": f"{title} > {sec_num} {sec_title} > {number}",
                    "sha256": sha256(rest),
                    "is_heading": False
                })
                i += 1
                continue

            # b) Bullet line under a prior list header
            b = BULLET_RE.match(raw)
            if b and current_list_number:
                bullet_index += 1
                number = f"{current_list_number}"
                text_ = compact(b.group(1))
                sec["clauses"].append({
                    "id": make_id(policy_id, number),
                    "number": number,
                    "title": current_list_title,
                    "text": "• " + text_,
                    "category": infer_category(current_list_number, sec_title, category_map),
                    "section_number": sec_num,
                    "path": f"{title} > {sec_num} {sec_title} > {current_list_number} {current_list_title}",
                    "sha256": sha256(text_),
                    "is_heading": False
                })
                i += 1
                continue

            # c) If we hit a non-bullet after finishing a list, reset list state
            if current_list_number and not b:
                # If a bullet appears soon, treat this as a subheading/context under the SAME list
                if _peek_has_bullet(sec_body, i, lookahead=10):
                    subtxt = compact(raw)
                    number = f"{current_list_number}.sub.{sha256(subtxt)[:6]}"
                    sec["clauses"].append({
                        "id": make_id(policy_id, number),
                        "number": number,
                        "title": current_list_title,   # keep the header title
                        "text": subtxt,                # the sub-heading line (e.g., "Privacy violations")
                        "category": infer_category(current_list_number, sec_title, category_map),
                        "section_number": sec_num,
                        "path": f"{title} > {sec_num} {sec_title} > {current_list_number} {current_list_title} > sub",
                        "sha256": sha256(subtxt),
                        "is_heading": False
                    })
                    i += 1
                    continue
                else:
                    # No bullets coming: close the list and re-process this line normally
                    current_list_number = None
                    current_list_title  = None
                    bullet_index = 0
                    # DO NOT i += 1 here; let the loop re-evaluate this line
                    continue


            # d) Plain paragraph (no number, no bullet)
            if raw:
                number = f"{sec_num}.context.{sha256(raw)[:6]}"
                sec["clauses"].append({
                    "id": make_id(policy_id, number),
                    "number": number,
                    "title": f"{sec_title} Context",
                    "text": raw,
                    "category": infer_category(sec_num, sec_title, category_map),
                    "section_number": sec_num,
                    "path": f"{title} > {sec_num} {sec_title} > Context",
                    "sha256": sha256(raw),
                    "is_heading": False
                })
            i += 1

        sections.append(sec)

    policy_doc = {
        "id": policy_id,
        "title": title,
        "version": version,
        "effective_on": date.today().isoformat(),
        "owner": owner,
        "uri": policy_uri,
        "sha256": sha256(text),
        "sections": sections
    }
    return policy_doc

# -------------------------
# Internals
# -------------------------
def _match_section(line: str):
    for rx in SECTION_RES:
        m = rx.match(line)
        if m:
            return m.group(1), m.group(2)
    # Fallback: treat first line as "1. <line>"
    return "1", line

def _peek_has_bullet(lines: List[str], start_idx: int, lookahead: int = 10) -> bool:
    seen = 0
    for j in range(start_idx, min(len(lines), start_idx + 30)):
        l = lines[j].strip()
        if not l:
            continue
        if BULLET_RE.match(l):
            return True
        seen += 1
        if seen >= lookahead:
            break
    return False

```

```python
records = []
for idx, file in files.iterrows():
    file_location = file['path']
    policy_id = file['POLICY_ID']

    title = file['Policy']
    source_framework = file["framework"]
    version = file['VERSION']
    owner = "Chief Technology Officer"
    policy_uri = f"blob://{file_location}"
    company_type = file.get("company_type", "")
    client_id = file['client_id']

    try:
        with open(file_location, "r", encoding="utf-8") as file:
            text = file.read()
    except UnicodeDecodeError:
        with open(file_location, "r", encoding="latin-1") as file:
            text = file.read()
                
    policy = parse_policy(
            text=text,
            policy_id=policy_id,
            title=title,
            version=version,
            owner=owner,
            policy_uri=policy_uri
        )

    policy_title = policy.get("title", "")
    
    policy_origin = policy.get("origin", "")

    for section in policy.get("sections", []):
        section_id = section.get("id", "")
        section_title = section.get("title", "")
        num = ''
        for item in section.get("clauses", []):
            text = ''
            print_num = item['number']
            full_line = ''
            if print_num == num:
                print_num = ''
                
            if item.get('is_heading'):
                full_line = print_num + ' ' + item['title']
                num = item['number']
                text = item['title']
            else:
                text = item['text']
                if 'sub' in item['number'] or 'context' in item['number']:
                    full_line = item['text']
                else:
                    full_line = print_num + ' ' + item['text']
                    num = item['number']

            records.append({
                "policy_id": policy_id,
                "policy_title": title,
                "source_framework": source_framework,
                "policy_origin": policy_origin,
                "section_id": section_id,
                "section_title": section_title,
                "is_heading": item.get("is_heading", False),
                "clause_id": item.get("id", ""),
                "clause_text": text,
                "clause_full_text": full_line,
                "clause_section_number": print_num,
                "company_type": company_type,
                "company_id": client_id,
            })
```

```python
def extract_numeric_section(val: str) -> str:
    val = str(val)
    val = val.strip()
    parts = val.split(".")
    numeric_parts = []
    
    for p in parts:
        if p.isnumeric():      # keep if it's a number
            numeric_parts.append(p)
        else:                  # stop at first non-numeric piece
            break
    
    return ".".join(numeric_parts)
```
